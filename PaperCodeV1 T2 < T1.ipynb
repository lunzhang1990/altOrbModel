```python
# permutations using library function 
from itertools import permutations 
  
# Get all permutations of [1, 2, 3] 
perm = permutations([i for i in range(8)])

def testPermuation(p):
    if not (p.index(4) < p.index(0) <  p.index(1)):
        return False
    
    if not (p.index(1) < p.index(6)):
        return False
    
    if not (p.index(2) < p.index(6)):
        return False
    
    if not (p[-1] == 7):
        return False
    
    if not (p.index(2) < p.index(3)):
        return False
    
    if not (p.index(3) < p.index(5)):
        return False
    
    if (p.index(3) < p.index(1)) and (p.index(6) < p.index(5)):
        return False
    
    if not (p.index(1) < p.index(5)):
        return False
    
    if (p.index(4) > p.index(2)) and (p.index(6) < p.index(3)):
        
        return False
    
    if (p.index(4) > p.index(2)) and (p.index(1) < p.index(3)):
        return False
    
    return True

ret21 = []
for p in perm:
    if testPermuation(p):
        ret21.append(p)
```

```python
def takeIntersection(s):
    ret = []
    for i in range(len(s[0])):
        t = [set(x[i]) for x in s]
        r = set()
        for tt in t:
            r |= set(tt)
        r = tuple(r)
        ret.append(r)
        
    return ret
```

```python
transBackOrder = []
for c in candidateElements:
    torder = takeIntersection(c)
    transBackOrder.append(torder)
```

```python
def insertTheta(order):
    ret = []
    for i in range(len(order)+1):
        order1 = list(order).copy()
        order1[i:i] = [9]
        for j in range(i+1, len(order1)+1):
            order2  = order1.copy()
            order2[j:j] = [8]
            ret.append(order2)
    return ret
```

```python
totalOrders = []
for order in ret21:
    totalOrders+= insertTheta(order)
len(totalOrders)
```

```python
def isContaining(poset,order):
    s1, s2, s3, s4 = set(), set(), set(), set()
    for i in order:
        if i in [8,9]:
            continue
        if order.index(i) < order.index(8):
            s1.add(i)
        if order.index(i) > order.index(8):
            s2.add(i)
        if order.index(i) < order.index(9):
            s3.add(i)
        if order.index(i) > order.index(9):
            s4.add(i)
    s = [s1, s2, s3, s4]
    #print(s)
    for i in range(len(poset)):
        if set(poset[i]).issubset(s[i]):
            continue
        else:
            return False
    return True
```

```python
def isContained(sub,sup):
    for i in range(len(sub)):
        if set(sub[i]).issuperset(set(sup[i])):
            continue
        else:
            return False
    return True

def identifyIndex8(l):
    d8 = ((1,),(),(0,),(1,))
    t8 = ((),(6,),(2,),())
    r8 = ((3,),(7,),(),(3,))
    s18 = ((),(),(4,),())
    s28 = ((5,),(),(),(5,))
    s38 = ((),(5,),(),())
    
    indexList = [d8,t8,r8,s18,s28,s38]
    portrait=[]
    for region in indexList:
        if isContained(l,region):
            portrait.append(1)
        else:
            portrait.append(0)           
    return tuple(portrait)
```

```python
# generate the prec and desc for each 0,..., 7 polynomials
partialOrderBase =[(4,0),(0,1),(1,6),(2,6),(1,7),(6,7),(5,7),(2,3),(3,5),(1,5),(2,7),(3,7),(4,7)] 
prec = dict()
desc = dict()
for i,j in partialOrderBase:
    if i not in desc:
        desc[i] = []
    desc[i].append(j)
    
    if j not in prec:
        prec[j] = []
    prec[j].append(i)
    
def genPrecDict(i,prec):
    
    if i not in prec:
        return set()
    
    ret = set(prec[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in prec:
                ret |= set(prec[j])
        
        if ret == preRet:
            return ret
        
def genDescDict(i,desc):
    
    if i not in desc:
        return set()
    
    ret = set(desc[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in desc:
                ret |= set(desc[j])
        
        if ret == preRet:
            return ret

preDict = dict()
desDict = dict()

for i in range(8):
    preDict[i] = genPrecDict(i,prec)
    desDict[i] = genDescDict(i,desc)
```

```python
# simplify the defining inequalities
def getMaixmal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in desDict:
            intersect = set(l)&set(desDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret

def getMinimal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in preDict:
            intersect = set(l)&set(preDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret
```

```python
def removeRedundant(l):
    ret = []
    
    for i in range(len(l)):
        flag = True
        for j in range(len(l)):
            if i!=j and isContained(l[i],l[j]):
                flag = False
                break
        if flag:
            ret.append(l[i])
    return ret
```

```python
def generatePairConnectedComponent(indexSet,atomicComponents):
    fUnion = []
    for i in range(len(indexSet)):
        curr = indexSet[i]
        toRemove = []
        for j in range(len(fUnion)):
            currUnion = fUnion[j]
            for k in range(len(currUnion)):
                if isConnected(indexSet[i],currUnion[k],atomicComponents):
                    #print(indexSet[i],currUnion[k])
                    toRemove.append(j)
                    break
        newUnion = [curr]
        for j in range(len(toRemove)):
            index = toRemove[j]
            newUnion+=fUnion[index]
        updatedUnion = []
        for j in range(len(fUnion)):
            if j not in toRemove:
                updatedUnion.append(fUnion[j])
        updatedUnion.append(newUnion)
        fUnion = updatedUnion
           
    return fUnion

def isOrderContained(order, po):
    for i in po[0]:
        if order.index(i) > order.index(8):
            return False
    
    for i in po[1]:
        if order.index(i) < order.index(8):
            return False
    
    for i in po[2]:
        if order.index(i) > order.index(9):
            return False
        
    for i in po[3]:
        if order.index(i) < order.index(9):
            return False
        
    return True

def isConnected(c1,c2, atomicComponents):
    intersection = takeIntersection([c1,c2])
    for atom in atomicComponents:
        if isOrderContained(atom,intersection):
            #print(atom)
            #print(intersection)
            return True
    return False
```

```python
def isAdjacentOrder(order1, order2):
    differ = []
    for i in range(len(order1)):
        if order1[i]!= order2[i]:
            differ.append(i)
            
    if len(differ)!= 2:
        return False
    
    if abs(differ[0] - differ[1]) == 1:
        return True
    else:
        return False

def isAdjacentOrders(orders1, orders2):
    for order1 in orders1:
        for order2 in orders2:
            if isAdjacentOrder(order1, order2):
                return True
    return False
```

```python
def isSample(r, po,theta1,theta2):
    for i in po[0]:
        if r[i] >= theta1:
            return False
    for i in po[1]:
        if r[i]  <= theta1:
            return False
        
    for i in po[2]:
        if r[i] >=theta2:
            return False
        
    for i in po[3]:
        if r[i]  <= theta2:
            return False
    return True

def generateCoveringIndex(parameter, CoveringSets):
    parameter[:2] = sorted(parameter[:2])[::-1]
    theta1, theta2, mu,beta, pi, d = parameter
    factor = (pi*pi + 2*pi*d)*mu
    #factor = 1

    dl2 = beta*(pi+d) + (1+beta)*d
    du2 = beta*d + (1+beta)*(pi+d)
    dl1 = beta*d + (1+beta)*(pi+d)
    tl2 = (beta*mu*(pi+d)+(2+2*beta+mu+beta*mu)*d)/(1+mu)
    tu1 = (beta*mu*d + (2+2*beta+mu+beta*mu)*(pi+d))/(1+mu)
    ru2 = ((mu+beta*mu)*(pi+d)+(2+2*beta+mu+beta*mu)*d)/(1+mu)
    rl1 = ((mu+beta*mu)*(pi+d)+(2+2*beta+mu+beta*mu)*d)/(1+mu)
    ru1 = ((mu+beta*mu)*d+(2+2*beta+mu+mu*beta)*(pi+d))/(1+mu)
    sl12 = beta*(2*d+pi)
    su22 = (1+beta)*(2*d+pi)
    sl21 = (1+beta)*(2*d+pi)
    su31 = (1+beta)*(2*d+pi)

    scaledList1 = [dl1,tu1,rl1,ru1,sl21,su31]
    scaledList2 = [dl2,du2,tl2,ru2,sl12,su22]

    theta1List = [x/factor for x in scaledList1]
    theta2List = [x/factor for x in scaledList2] 
    
    r = theta2List + [theta1List[1], theta1List[3]]
    
    
    for po in CoveringSets:
        if isSample(r,po,theta1,theta2):
            return (po,r)
        
    return None
```
